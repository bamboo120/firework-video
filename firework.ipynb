```python
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#load data
dx_test=pd.read_csv('dx.csv')
dy_test=pd.read_csv('dy.csv')
dxdy_train=pd.read_csv('dxdy2.csv')
A_test=pd.read_csv('intensity.csv')
A_train=pd.read_csv('intensity2.csv')
out_test=pd.read_csv('number.csv')
out_train=pd.read_csv('number2.csv')
A_test = A_test.join(dx_test, how='outer')
A_test = A_test.join(dy_test, how='outer')
A_train = A_train.join(dxdy_train, how='outer')
Y_test = pd.read_csv('number.csv')
Y_train = pd.read_csv('number2.csv')

A_test['R_I_add']=A_test['R_I_add']-A_test['R_I_add'][0]
A_test['G_I_add']=A_test['G_I_add']-A_test['G_I_add'][0]
A_test['B_I_add']=A_test['B_I_add']-A_test['B_I_add'][0]

A_train['R_I_add']=A_train['R_I_add']-A_train['R_I_add'][0]
A_train['G_I_add']=A_train['G_I_add']-A_train['G_I_add'][0]
A_train['B_I_add']=A_train['B_I_add']-A_train['B_I_add'][0]

```

```python
clf = Ridge(alpha=10.0,normalize=True)
clf.fit(A_train,Y_train)
```

```python
Y_predict = clf.predict(A_test)
```

```python
Y_predict = Y_predict-Y_predict[0]
l = len(Y_predict)
x = list(range(0,l))
```

```python
plt.plot(x,Y_predict, 'g--', linewidth=2,label = 'Prediction')
plt.plot(x,Y_test, 'r-', linewidth=2,label = 'Actual')
plt.title('predicted firework number by video vs actual')
plt.xlabel('time [frame]')
plt.ylabel('firework number')
plt.legend()
```
